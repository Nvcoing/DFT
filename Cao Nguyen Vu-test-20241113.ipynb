```python
import numpy as np
from scipy.stats import norm, poisson, chi2_contingency
from numpy.linalg import det, eig
```

#### Câu 1:

##### a) Xác suất để chiều cao của một sinh viên ngẫu nhiên nằm trong khoảng từ 165 cm đến 175 cm:

Vì chiều cao tuân theo phân phối chuẩn với 𝜇 = 170 và 𝜎 = 5, ta tính xác suất:  

```python
u = 170
sigma = 5
p_165_175 = norm.cdf(175, u, sigma) - norm.cdf(165, u, sigma)
print(f"Xác suất chiều cao nằm trong khoảng từ 165 cm đến 175 cm: {p_165_175:.4f}")
```

##### b) Xác suất để chiều cao của một sinh viên cao hơn 180 cm:

```python
p_above_180 = 1 - norm.cdf(180, u, sigma)
print(f"Xác suất chiều cao cao hơn 180 cm: {p_above_180:.4f}")
```

#### Câu 2:

##### a) Xác suất để nhà hàng nhận đúng 8 yêu cầu trong một giờ cao điểm (phân phối Poisson với (lambda = 5):


```python
lam = 5
p_8_requests = poisson.pmf(8, lam)
print(f"Xác suất nhận đúng 8 yêu cầu: {p_8_requests:.4f}")
```

##### b) Xác suất để nhận ít hơn 3 yêu cầu trong một giờ cao điểm:
    

```python
p_less_than_3 = poisson.cdf(2, lam)
print(f"Xác suất nhận ít hơn 3 yêu cầu: {p_less_than_3:.4f}")
```

#### Câu 3:

##### Kiểm định Chi-Square để kiểm tra sự khác biệt về lượng hàng hóa bán ra vào các ngày trong tuần:


```python
# Tạo bảng dữ liệu quan sát
data = np.array([
    [20, 30, 25],  # Thứ Hai
    [25, 28, 27],  # Thứ Ba
    [22, 35, 30],  # Thứ Tư
    [30, 25, 32]   # Thứ Năm
])

# Thực hiện kiểm định Chi-Square
chi2_stat, p_value, dof, expected = chi2_contingency(data)
alpha = 0.05
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"p-value: {p_value:.4f}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:\n", expected)
if p_value < alpha:
    print("Có sự khác biệt đáng kể về lượng hàng hóa bán ra vào các ngày trong tuần(bác bỏ giả thuyết null).")
else:
    print("Không có sự khác biệt đáng kể về lượng hàng hóa bán ra vào các ngày trong tuần(không đủ bằng chứng bác bỏ giải thuyết null).")
```

#### Câu 4:

```python
A = np.array([[2, 3], [-1, 4]])
B = np.array([[1, -2], [0, 5]])

A_plus_B = A + B
A_minus_B = A - B
print("A + B:\n", A_plus_B)
print("A - B:\n", A_minus_B)
```

```python
AB = np.dot(A, B)
BA = np.dot(B, A)
print("AB:\n", AB)
print("BA:\n", BA)
```

```python
det_A = det(A)
det_B = det(B)
print(f"Determinant of A: {det_A:.4f}")
print(f"Determinant of B: {det_B:.4f}")
```

#### Câu 5:

##### a) Tìm các giá trị riêng của ma trận 

```python
A = np.array([[4, 1], [2, 3]])
eigenvalues, eigenvectors = eig(A)
print("Eigenvalues of A:", eigenvalues)
```

##### b) Tìm các vector riêng tương ứng với mỗi giá trị riêng tìm được:

```python
print("Eigenvectors of A:\n", eigenvectors)
```

##### c) Xác minh rằng các vector riêng thỏa mãn phương trình

```python
for i in range(len(eigenvalues)):
    lambda_i = eigenvalues[i]
    v_i = eigenvectors[:, i]
    Av = np.dot(A, v_i)
    lambda_v = lambda_i * v_i
    print(f"Kiểm tra cho eigenvalue {lambda_i:.4f}:")
    print("A * v:", Av)
    print("lambda * v:", lambda_v)
    print("Kiểm tra:", np.allclose(Av, lambda_v), "\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Tính toán các xu hướng trung tâm
mean_values = data[['Open', 'High', 'Low', 'Close']].mean()
median_values = data[['Open', 'High', 'Low', 'Close']].median()
mode_values = data[['Open', 'High', 'Low', 'Close']].mode().iloc[0]

# Thiết lập biểu đồ
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Phân tích thống kê các giá vàng', fontsize=16)

# Biểu đồ cột cho trung bình, trung vị, và mode
axes[0, 0].bar(mean_values.index, mean_values.values, color='blue', label='Mean')
axes[0, 0].bar(median_values.index, median_values.values, color='green', alpha=0.7, label='Median')
axes[0, 0].bar(mode_values.index, mode_values.values, color='red', alpha=0.5, label='Mode')
axes[0, 0].set_title('Xu hướng trung tâm (Mean, Median, Mode)')
axes[0, 0].legend()

# Biểu đồ boxplot và violin plot
sns.boxplot(data=data[['Open', 'High', 'Low', 'Close']], ax=axes[0, 1])
axes[0, 1].set_title('Box Plot cho Open, High, Low, và Close')
axes[0, 1].set_ylabel('Giá')

sns.violinplot(data=data[['Open', 'High', 'Low', 'Close']], ax=axes[1, 0])
axes[1, 0].set_title('Violin Plot cho Open, High, Low, và Close')
axes[1, 0].set_ylabel('Giá')

# Biểu đồ phân phối giá (Histogram)
for column in ['Open', 'High', 'Low', 'Close']:
    sns.histplot(data[column], bins=30, kde=True, ax=axes[1, 1], label=column, element='step')
axes[1, 1].set_title('Biểu đồ phân phối giá (Histogram)')
axes[1, 1].legend()

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
